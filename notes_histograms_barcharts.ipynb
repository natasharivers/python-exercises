```python
import matplotlib.pyplot as plt
import math
import random
```

```python
plt.figure(figsize = (10,6))

x1 = [random.gauss(0, 0.8) for _ in range(1000)]
x2 = [random.gauss(-2, 1) for _ in range(1000)]
x3 = [random.gauss(3, 1.5) for _ in range(1000)]
x4 = [random.gauss(7, 0.6) for _ in range(1000)]

#_ is placeholder variable.

kwargs = dict(alpha = 0.3, bins =25, density = True)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
plt.hist(x4, **kwargs)

plt.show()
```

```python
#make list of lists to iterate

x= [x1,x2,x3,x4]

for i in range(1,5):
    plt.subplot(2,2,i)
    plt.hist(x[i-1], color = 'white', edgecolor = 'k')
```
