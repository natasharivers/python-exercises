```python
import matplotlib.pyplot as plt
import math
```

```python
x = range(-50, 51)
y = [n ** 2 for n in x]
plt.scatter(x, y)
plt.title('$y = x^2$', fontsize = 14)
plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$x^2$', fontsize = 14)
plt.xticks([-50,-40,-30,-20,-10,0,10,20,30,40,50], rotation = 45)
#^--- ticks at bottom of plot and rotation of x axis
plt.show()
```

```python
#define an x:
x = [x * .25 for x in range(-25,25)]

y = [math.sin(x) for x in x]
y2 = [math.cos(x) for x in x]
y3 = [math.sin(x) * -1 for x in x]
y4 = [math. cos(x) * -1 for x in x]
```

```python
y #"big ol list of float values"
```

```python
#define a figure
pi = math.pi

plt.figure(figsize = (10,8))

plt.suptitle('Sine and Cosine Functions', fontsize = 16) 
#^-- overall title for series of subplots

#graph #1:
plt.subplot(2,2,1)
#plt.subplot(length, hieght, plot being represented)
plt.plot(x,y, ls= '-', marker = '*', color ='red')
plt.title('$sin(x)$')
plt.grid(True, ls = '--')
#xtick
plt.xticks([-2 * pi, -1 * pi, 0, pi, 2 * pi])

#graph #2:
plt.subplot(2,2,2)
plt.plot(x, y2, marker = 's', color = 'blue', ls = '')
plt.title('$cos(x)$')
plt.axhline(0, color = 'k', linewidth = 0.5) #horizonal axis line

#graph #3:
plt.subplot(2,2,3)
plt.plot(x, y3, marker = 'v', color = 'green', ls = '--')
plt.title('$-sin(x)$')
plt.ylabel('y 3 label')

#graph #4:
plt.subplot(2,2,4)
plt.plot(x, y4, color = 'm', marker = 'o')
plt.title('$-cos(x)$')
plt.axhline(0, color = 'k', linewidth = 0.5)
plt.fill_between(x, y4, color ='m')



plt.show()
```
