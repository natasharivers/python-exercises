```python
for num in range(1,101):
    string = ""
    if num % 3 == 0:
        string = string + "Fizz"
    if num % 3 != 0:
        string = string + str(num)
    print(string)
```

```python
for num in range(1,101):
    string = ""
    if num % 5 == 0:
        string = string + "Buzz"
    if num % 5 != 0:
        string = string + str(num)
    print(string)
```

```python
for num in range(1,101):
    string = ""
    if num % 5 == 0:
        string = string + "Buzz"
    if num % 3 == 0:
        string = string + "Fizz"
    elif num % 5 != 0 and num % 3 != 0 :
        string = string + str(num)
    print(string)
```

```python
for num in range(1, 101):
    if num % 3 == 0:
        print('Fizz')
    if num % 5 == 0:
        print('Buzz')
    if num % 3 == 0 and num % 5 == 0:
        print('FizzBuzz')
    else:
        print(num)
```

```python
for num in range(1, 50):
    if num % 3 == 0:
        print('Fizz')
    if num % 5 == 0:
        print('Buzz')
    if num % 3 == 0 and num % 5 == 0:
        print('FizzBuzz')
    else:
        print(num)
```

```python
def squared(i):
    return(i*i)
def cubed(i):
    return(i*i*i)
```

```python
integer = int(input("what number would you like to go up to?"))
print("number       squared     cubed  ")
print("---------    ---------   --------")

for i in range(integer):
    print(i,"\t", squared(i), "\t", cubed(i))
```
