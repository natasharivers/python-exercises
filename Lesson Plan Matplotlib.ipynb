```python
import matplotlib.pyplot as plt
```

```python
x = list(range(150))
plt.plot(x)
plt.show()
```

```python
plt.plot(range(len(x)), x)
plt.show()
```

```python
plt.plot(x)
plt.xlim(-20, 200)
plt.ylim(-5, 160)
plt.show()
```

```python
plt.plot(x, c= 'red')
plt.show()
```

```python
from random import randint
```

```python
x1= [randint(-5,25) + n for n in range(150)]
x2= [x + randint(-6,6) for x in x1]
```

```python
plt.plot(x1, c= 'green', alpha=0.6)
plt.plot(x2, c='red', alpha = 0.4)
plt.show()
#line plot
```

```python
plt.scatter(range(len(x1)), x1)
plt.scatter(range(len(x2)), x2)
plt.show()
```

```python
plt.scatter(range(len(x1)), x1, s=30, c='red')
plt.scatter(range(len(x2)), x2, s=10, c= 'green')
plt.title('A couple of random series')
plt.show()
```

```python
x = list(range(-49, 50))
y = [n ** 2 for n in x]
plt.scatter(x, y)
plt.title('A Quadratic Distribution, $y = x^2$')
plt.xlabel('$x$')
plt.ylabel('$x^2$')
plt.show()
#labeling x and y axis
```

```python
x = [1, 2, 3.14, 4]
y = [n ** 2 for n in x]
plt.scatter(x, y)
plt.xticks([1, 2, 3.14, 4], ['one', 'two', '$\pi$', 'four'])
plt.show()
#add custom ticks
```

```python
import math

x = range(-4, 4)
y = [math.sin(n) for n in x]
plt.scatter(x, y)
plt.plot(x, y, ls=':')
plt.title('An approximation of the sine function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
#linestyle added
```
